```python
da.rename(columns = lambda x:x.replace(':',''), inplace = True)
#showcols(da)
#da.groupby('Nature').count()['id']
da.loc[da['Nature']=='Unknown (UNK)','Nature'] = 'Unknown'
#Unknown merged.
da.drop(da[da['Phase'] == 'Standing (STD)'].index, inplace = True) # Stationary Data were dropped
da.drop(da[da['Phase'] == 'Unknown (UNK)'].index, inplace = True)
da.reset_index()
```

```python
### Crew Cleaning
#showcols(da)
da.groupby('Crew').count()['id']
da.drop(da[da['Crew'] == 'Fatalities:  / Occupants:'].index,inplace = True) # Drop unrecorded columns
da.groupby('Crew').count()['id']
showcols(da)
da.drop(da.columns[[6,7,8,20,22,23]],axis = 1, inplace=True)
```

```python
'''Dropped plane Types which are nulls.'''
showcols(da)
da.drop(da[da['Type'].isnull()].index, inplace = True)
#da[da['Type'].str.contains('Boeing',case = False)].groupby('Type').count()['id'].nlargest(25)
```

```python
#Date formatting
da.drop(da[da['Date'].str.contains('xx',case = False)].index,inplace = True)
da['Raw_Time'] = da['Date'].apply(lambda x:time.strptime(x,"%A %d %B %Y"))
da['Raw_Time'].head(5)
da.reset_index()
```

```python
da['Elevations'] = [None for i in range(len(da['Total_Fatality']))]
showcols(da)
da.head(3)
da.reset_index()
da['Total_Fatality'] = [get_occupants(i)[0] for i in da['Total']]
da['Total_Occupants'] = [get_occupants(i)[1] for i in da['Total']]
da['Fatality_Rate'] = da['Total_Fatality']/da['Total_Occupants']
da['Fatality_Rate'].describe()
da.drop(da[da['Fatality_Rate']>1].index,inplace = True)
da['Fatality_Rate'].describe()
da[da['Fatality_Rate'] ==max(da['Fatality_Rate'])]
```

```python
#Group by make and types
da['Type'] = da['Type'].apply(lambda x:x.capitalize())
da['Type']
da['Make'] = da['Type'].apply(lambda x:' '.join(x.split()[1:]))
da['Type'] = da['Type'].apply(lambda x:x.split()[0])
da.groupby('Type').mean()['Fatality_Rate'].nsmallest(50)
```
